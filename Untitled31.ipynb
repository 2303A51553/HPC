```python
import matplotlib.pyplot as plt

conditions = ["Clear", "Rain", "Fog", "Night"]
accuracy = [95, 88, 82, 85]

plt.figure()
plt.bar(conditions, accuracy)
plt.xlabel("Driving Conditions")
plt.ylabel("Detection Accuracy (%)")
plt.title("AI Detection Accuracy Under Different Weather Conditions")
plt.show()
```

```python
import matplotlib.pyplot as plt

sensors = ["Camera", "LiDAR", "Radar"]
reliability = [85, 92, 89]

plt.figure()
plt.bar(sensors, reliability)
plt.xlabel("Sensor Type")
plt.ylabel("Reliability Score (%)")
plt.title("Reliability Comparison of Autonomous Vehicle Sensors")
plt.show()
```

```python
import matplotlib.pyplot as plt

causes = ["Sensor Failure", "Weather Conditions", "Edge Cases", "Software Bugs"]
values = [30, 25, 25, 20]

plt.figure()
plt.pie(values, labels=causes, autopct='%1.1f%%')
plt.title("Major Causes of AI System Failures")
plt.show()
```
