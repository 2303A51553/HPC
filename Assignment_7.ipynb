#1.Vector Addition (Scalar vs SIMD-like)

```python
import numpy as np
import time

N = 10_000_000

A = np.arange(N, dtype=np.float64)
B = np.arange(N, dtype=np.float64)

# Scalar loop
C = np.zeros(N)
start = time.time()
for i in range(N):
    C[i] = A[i] + B[i]
end = time.time()
print("Normal loop time:", end - start)

# Vectorized (SIMD-like)
start = time.time()
C = A + B
end = time.time()
print("Vectorized time:", end - start)
```

#Explanation:

* This code compares **Python loop vs NumPy vectorization**
* Normal loop adds numbers **one by one** → **slow**
* `C = A + B` adds **all elements at once** → **very fast**
* NumPy uses **optimized C code + CPU SIMD**
* **Result:** Vectorized code is much faster than loop


#2.Reduction (Sum)

```python
import numpy as np
import time

N = 10_000_000
A = np.ones(N, dtype=np.float64)

# Normal loop
start = time.time()
s = 0.0
for i in range(N):
    s += A[i]
end = time.time()
print("Normal sum:", s, "Time:", end - start)

# Vectorized reduction
start = time.time()
s = np.sum(A)
end = time.time()
print("Vectorized sum:", s, "Time:", end - start)
```

#Explanation:

* This code compares **manual sum vs NumPy sum**
* Loop adds elements **one by one** → **slow**
* `np.sum(A)` sums **all elements at once**
* NumPy uses **optimized C + SIMD**
* **Result:** Vectorized sum is much faster


#3.Memory Alignment Effect

```python
import numpy as np
import time

N = 10_000_000

unaligned = np.arange(N + 1, dtype=np.float64)[1:]
aligned = np.arange(N, dtype=np.float64)

start = time.time()
np.sum(unaligned)
print("Unaligned time:", time.time() - start)

start = time.time()
np.sum(aligned)
print("Aligned time:", time.time() - start)
```

#Explanation:



* `aligned` → data starts at proper memory boundary
* `unaligned` → data starts at offset (misaligned)
* `np.sum()` on **aligned data is faster**
* Unaligned memory makes CPU work harder
* **Result:** Aligned arrays give better performance



#4.Parallel + SIMD (Implicit)

```python
import numpy as np
import time

N = 10_000_000
A = np.arange(N, dtype=np.float64)

start = time.time()
B = A * 2.0
print("Vectorized (SIMD + multithreaded) time:", time.time() - start)
```

#Explanation



* Creates a large NumPy array `A`
* `B = A * 2.0` multiplies **all elements at once**
* NumPy uses **SIMD + multiple CPU threads**
* No Python loop → very fast
* Printed time shows **optimized vectorized performance**


#5. Branch Divergence

```python
import numpy as np
import time

N = 10_000_000
A = np.random.rand(N) * 100
B = np.zeros(N)

start = time.time()
for i in range(N):
    if A[i] > 50:
        B[i] = A[i] * 2
    else:
        B[i] = A[i] / 2
print("Branch loop time:", time.time() - start)
```

```python
start = time.time()
B = np.where(A > 50, A * 2, A / 2)
print("Vectorized conditional time:", time.time() - start)
```

#Explanation:



* Loop checks condition **one by one** → slow (branching)
* `np.where()` applies condition **to all elements at once**
* NumPy avoids Python branching
* Uses **vectorized + SIMD** operations
* **Result:** Vectorized conditional is much faster
